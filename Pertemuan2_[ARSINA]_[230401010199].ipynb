```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
# Jawaban Q1
print(f"Jumlah Baris: {df.shape[0]}, Jumlah Kolom: {df.shape[1]}")
print("Nama Kolom:", df.columns.tolist())
```

```python
# Jawaban Q2
missing_data = df.isnull().sum()
proporsi = (missing_data / len(df)) * 100
df_missing = pd.DataFrame({'Jumlah Missing': missing_data, 'Persen (%)': proporsi})
print(df_missing[df_missing['Jumlah Missing'] > 0])
```

```python
# Jawaban Q3
persen_selamat = df['Survived'].mean() * 100
print(f"Persentase Penumpang Selamat: {persen_selamat:.2f}%")
```

```python
# Jawaban Q4
wanita_kelas1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(f"Jumlah Wanita Kelas 1: {len(wanita_kelas1)}")
print(f"Rata-rata Usia: {wanita_kelas1['Age'].mean():.2f} tahun")
```

```python
# Jawaban Q5
tingkat_selamat = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print(tingkat_selamat)
```

```python
# Jawaban Q6
analisis_gender = df.groupby('Sex')['Survived'].mean()
print(analisis_gender)
```
